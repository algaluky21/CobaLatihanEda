```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.shape
```

```python
iris.head()
```

```python
iris.plot(x='nokia', y='samsung')
```

```python
iris.plot(x='nokia', y='samsung', kind='scatter')
plt.xlabel('sepal nokia (cm)')
plt.ylabel('sepal samsung (cm)')
```

```python
iris.plot(y='nokia', kind='box')
plt.ylabel('samsung (cm)')
```

```python
iris.plot(y='nokia', kind='hist')
plt.xlabel('nokia (cm)')
```
